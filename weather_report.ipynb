```python
def create_predictions(predictors,core_data,reg):
    train=core_data.loc[:"2024-06-30"]
    test=core_data.loc["2024-07-01":]
    reg.fit(train[["PRCP","TMAX","TMIN"]],train["Target"])
    predictions = reg.predict(test[["PRCP","TMAX","TMIN"]])
    error=mean_absolute_error(test["Target"],predictions)
    combined =pd.concat([test["Target"],pd.Series(predictions,index=test.index)],axis=1)
    combined.columns = ["Actual","Predictions"]
    return error,combined
```

```python
core_data["Month_max"]=core_data["TMAX"].rolling(30).mean()
```

```python
core_data
```

```python
error,combined=create_predictions(predictors,core_data,reg)
```

```python
predictors=["PRCP","TMAX","TMIN","Target","Month_max"]
```

```python
error,combined=create_predictions(predictors,core_data,reg)
```
